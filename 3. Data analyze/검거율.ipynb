```python
import pandas as pd
df = pd.read_excel("Data analyze\Datafile\관서별 5대범죄 발생 및 검거.xlsx")
```

```python
df = df.set_index('관서명')
df["구별"] = {'계': '구 없음', '서대문서': '서대문구', '수서서': '강남구', '강서서': '강서구', '서초서': '서초구',
'서부서': '은평구', '중부서': '중구', '종로서': '종로구', '남대문서': '중구',
'혜화서': '종로구', '용산서': '용산구', '성북서': '성북구', '동대문서': '동대문구',
'마포서': '마포구', '영등포서': '영등포구', '성동서': '성동구', '동작서': '동작구',
'광진서': '광진구', '강북서': '강북구', '금천서': '금천구', '중랑서': '중랑구',
'강남서': '강남구', '관악서': '관악구', '강동서': '강동구', '종암서': '성북구', 
'구로서': '구로구', '양천서': '양천구', '송파서': '송파구', '노원서': '노원구', 
'방배서': '서초구', '은평서': '은평구', '도봉서': '도봉구'}

#관서명마다 해당된 지역 매핑.
```

```python
df = pd.pivot_table(df, index='구별')   #'구별'을 인덱스로 설정, 본래 인덱스인 관서명은 사라짐.
```

```python
df = df.drop(['구 없음'])   #계는 필요가 없으므로, 구 없음을 인덱스로 가진 계 행은 삭제.
```

```python
df["강간검거율"] = df["강간(검거)"] / df["강간(발생)"] * 100
df["강도검거율"] = df["강도(검거)"] / df["강도(발생)"] * 100
df["살인검거율"] = df["살인(검거)"] / df["살인(발생)"] * 100
df["절도검거율"] = df["절도(검거)"] / df["절도(발생)"] * 100
df["폭력검거율"] = df["폭력(검거)"] / df["폭력(발생)"] * 100
df["검거율"] = df["소계(검거)"] / df["소계(발생)"] * 100

#검거율 = 검거 / 발생, 백분율의 표현을 위해서 100을 곱함.
```

```python
del df["강간(검거)"]
del df['강도(검거)']
del df['살인(검거)']
del df['절도(검거)']
del df['폭력(검거)']
del df['소계(발생)']
del df['소계(검거)']

#필요 없는 요소 삭제
```

```python
df.rename(columns={'강간(발생)':'강간', '강도(발생)':'강도', '살인(발생)':'살인', '절도(발생)':'절도', '폭력(발생)':'폭력'}, inplace=True)
df
```

```python
df_2 = pd.read_csv("Data analyze\Datafile\관서별 5대범죄 발생 및 검거.xlsx")
df_2 = df_2.set_index('구별')   #'구별'을 인덱스로 설정. 뒤이어 하는 병합을 위해 df와 같은 인덱스. 
df
```

```python
merged = df.join(df_2) #df에 df_2 join. 인덱스 같으므로 '구별'을 기준으로 병합됨.
```
